## **Conditions and Branching**

### **1. if Statement:**

↪  The if statement is used to make decisions in your code. It allows you to execute a block of code only if a certain condition is true.



```
if condition:
    # code to execute if condition is True

```

*   Condition: A boolean expression that is either True or False.
*   If the condition is True, the code inside the if block runs.
*   If the condition is False, the code inside the if block is skipped.


```
x = 10
if x > 5:
    print("x is greater than 5")

Output:
x is greater than 5
(Printed because the condition x > 5 is True.)

```



### **2. else Statement:**

↪  The else statement follows an if statement and runs a block of code if the if condition is False.



```
#syntax

if condition:
    # code to execute if condition is True
else:
    # code to execute if condition is False


```

```
#Example:

x = 3
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")

Output:
x is not greater than 5
(Printed because the condition x > 5 is False.)

```





### **3. elif (else if) Statement:**

↪  The elif (short for "else if") statement allows you to check multiple conditions. It comes after the if statement and before the else statement.



```
#syntax

if condition1:
    # code to execute if condition1 is True
elif condition2:
    # code to execute if condition2 is True
else:
    # code to execute if both conditions are False


```

```
#Example:

x = 7

if x > 10:
    print("x is greater than 10")
elif x == 7:
    print("x is 7")
else:
    print("x is less than or equal to 10 and not 7")



Output:
x is 7
(Printed because x == 7 is True.)

```

````python
# Input score
score = 85

# Determine grade
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

# Output the grade
print(f"The grade is: {grade}")

````

## **f in print**

↪  The f in print(f"Hello, {name}!") is used to create an f-string (formatted string).


↪  f: Tells Python, "I want to put the value of variables inside this string."


### **Why Use f?**

↪ Purpose: The f allows you to directly insert (or format) variables and expressions inside a string.

↪  How It Works: When you put f before the string, Python knows to replace anything inside {} with the value of the variable or expression.


###**Without f:**

↪  If you don't use f, you can't directly insert variables into the string. For example:


```
name = "Alice"
print("Hello, {name}!")  # This will literally print {name}, not Alice.

```



###**With f:**

↪  When you use f, Python replaces {name} with the value of name.


```
print(f"Hello, {name}!")  # This will correctly print: Hello, Alice!


```

##**Input in Python:**

↪  Purpose: The input() function is used to take user input in Python.


```
name = input("Enter your name: ")
print(name)

Output:
The program prompts: Enter your name:
If the user types Rajveer, it will print: Rajveer

```

###**Important Note:**


↪  By default: The input() function takes the input as a string.

↪ Converting Input to Other Types

↪ f you need the input as an integer or float, you must convert it using functions like int() or float().



```
age = int(input("Enter your age: "))
marks = float(input("Enter your marks: "))

The int() converts the input into an integer.
The float() converts the input into a floating-point number.
```
