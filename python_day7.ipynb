## **Reading Files with open()**

↪  you can read files using the open() function.

↪ This function allows you to access the content of a file, either to read from it, write to it, or append data.

### **Steps to Read a File**

*   Open the File: Use the open() function to open the file. You specify the file name and the mode ('r' for reading).

*   Read the Content: Use a method like read(), readline(), or readlines() to get the content.


*   Close the File: After you’re done, it’s good practice to close the file using the close() method.


```

Example: Reading the Entire File

# Open the file in read mode ('r')
file = open('example.txt', 'r')

# Read the content of the file
content = file.read()

# Print the content
print(content)

# Close the file
file.close()


Explanation:
* open('example.txt', 'r'): Opens the file example.txt in read mode.
* If the file doesn’t exist, it will raise an error.
* file.read(): Reads the entire content of the file.
* file.close(): Closes the file to free up resources.

```






###Example: Reading Line by Line
```


# Open the file in read mode ('r')
file = open('example.txt', 'r')

# Read and print each line one by one
for line in file:
    print(line, end='')  # 'end' is used to avoid adding extra newline

# Close the file
file.close()



Explanation:
* open('example.txt', 'r'): Opens the file example.txt in read mode.
* If the file doesn’t exist, it will raise an error.
* file.read(): Reads the entire content of the file.
* file.close(): Closes the file to free up resources.

```


##**Using with to Open Files**


↪  A more common and recommended approach is to use a with statement. This automatically closes the file for you:

```
Example:

# Open and read the file using 'with'
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)


Explanation:
with open(..., 'r') as file: Opens the file and assigns it to file.
When the block inside with is done, the file is automatically closed,
even if an error occurs.

```


## **Writing Files:**

↪  Used to add new content to a file or modify existing content.

↪ It either creates a new file or changes an existing one.

###**Writing ('w'):**

↪ Opens the file for writing

↪ if the file exists, its content is overwritten.

↪ If the file doesn’t exist, it creates a new one


```
with open('example.txt', 'w') as file:
    file.write("Hello, World!")
```
